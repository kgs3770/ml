```python
import pandas as pd

df = pd.read_csv('data/Fish.csv')
```

```python
perch_df = df.loc[df['Species'] == 'Perch']

perch_length = perch_df[['Length2']]
perch_weight = perch_df[['Weight']]
```

```python
#데이터 나눔
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
#기울기 or 가중치
lr.coef_
```

```python
#y축
lr.intercept_
```

```python
import matplotlib.pyplot as plt

plt.scatter(train_input, train_target)

#선그래프 그리기
plt.plot([15, 50], [15 * 37 - 667, 50 * 37 - 667])
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
#100cm짜리 예측해봐
lr.predict([[100]])
```

```python
#다항 회귀
#그래프를 곡선의 형태로 만들거임
```

```python
#학습 하려는 데이터를 제곱 후 train input에 넣어줌
train_input['Length2_poly'] = train_input['Length2'] ** 2
test_input['Length2_poly'] = test_input['Length2'] ** 2
```

```python
lr = LinearRegression()
lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.predict([[2, 4]]) #[x, x**2]
```

```python
print(lr.coef_, lr.intercept_)
```
